```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df_sample = df.head(365)

chuva = df_sample['02346002']
data = df_sample['DataHora']

#Plotando o grafico de linha
plt.figure(figsize=(10,5))
plt.plot(data,chuva,label='Série Temporal')

#Adicionando a media e o desvio padrao
media_serie = chuva.mean()
desvio_padrao = chuva.std()
plt.axhline(media_serie, color='r', linestyle='--', label=f'Média: {media_serie:.2f}')
plt.fill_between(data, media_serie - desvio_padrao, media_serie + desvio_padrao, color='b', alpha=0.2, label='Desvio padrão')

# Configurando o formato das datas no eixo x
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())  # Mostra um rótulo por mês

# Adicionando labels e legenda
plt.xlabel('Data')
plt.ylabel('Precipitação (mm)')
plt.title('Registros Estação 02346002 ')
plt.legend()
plt.show()

```
